```python
import pandas as pd
import numpy as np
game_file = "../Resources/Purchase_data.csv"
```

```python
game_df=pd.read_csv(game_file)
game_df.head()
```

```python
u_player=game_df['SN'].nunique()
avg_price=game_df['Price'].mean()
total_purchase=len(game_df['Purchase ID'])
total_rev=game_df['Price'].sum()
```

```python
sum_game={'Unique Player Count':u_player, 'Average Purchase Price':avg_price,'Number of Purchase':total_purchase,'Total Revenue':total_rev}
sum_game
summary_df=pd.DataFrame([sum_game])
summary_df
```

```python
group_gender=game_df.groupby('Gender')
gender_count=group_gender['SN'].nunique()
gen_percent=gender_count/u_player*100
sum_gender=pd.DataFrame({'Gender Count':gender_count,'Gender Percentage':gen_percent})
sum_gender
count_purchase=group_gender['Purchase ID'].count()
avg_pur_price=group_gender['Price'].mean()
avg_total_pur=group_gender['Price'].sum()
avg_pur_person=avg_total_pur/gender_count*100
gender_pur_sum=pd.DataFrame({'Purchase Count':count_purchase,'Average Purchase Price':avg_pur_price,'Average Purchase':avg_total_pur,'Average Purchase Per Person': avg_pur_person})
gender_pur_sum
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

game_df['Age Group']=pd.cut(game_df['Age'],bins,labels=labels)
game_df
group_age=game_df.groupby('Age Group')
count_age=group_age['SN'].nunique()
percent_age=count_age/u_player*100
age_sum=pd.DataFrame({'Players in Age Group':count_age,'Percent of Age Group':percent_age})
age_sum
```

```python
pur_ct_age=group_age["Purchase ID"].count()
avg_p_ct_age= group_age["Price"].mean()
tot_pur_age= group_age["Price"].sum()
avg_pur_per_age= tot_pur_age/count_age
age_pur_sum= pd.DataFrame({"Purchase Count": pur_ct_age,"Average Purchase Price": avg_p_ct_age,"Total Purchase Value":tot_pur_age,"Average Purchase Total Per Person": avg_pur_per_age})
age_pur_sum
```

```python
group_sn=game_df.groupby('SN')
pur_ct_sn=group_sn['Purchase ID'].count()
avg_pur_price=group_sn['Price'].mean()
tot_pur_amt=group_sn['Price'].sum()
top_sn_sum=pd.DataFrame({'Purchase Count':pur_ct_sn,'Average Purchase Price':avg_pur_price,'Total Purchase Amount':tot_pur_amt})
top_sn_sum=top_sn_sum.sort_values('Total Purchase Amount',ascending=False)
top_sn_sum.head(1)

# revisit how to list in a table
```

```python
### Most Popular Items

* Identify the 5 most popular items by purchase count, then list (in a table):
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value

Most Popular Items

    Retrieve the Item ID, Item Name, and Item Price columns

    Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value

    Create a summary data frame to hold the results

    Sort the purchase count column in descending order

    Optional: give the displayed data cleaner formatting

    Display a preview of the summary data frame

```

```python
item_df=game_df[["Item ID","Item Name","Price"]]
group_item=item_df.groupby(["Item ID","Item Name"])
pur_ct_item=group_item['Price'].count()
tot_item_val=group_item['Price'].sum()
avg_item_price=tot_item_val/pur_ct_item

hot_item=pd.DataFrame({'Purchase Count':pur_ct_item,'Average Item Price':avg_item_price,'Total Purchase Value':tot_item_val})
hot_item
hot_item=hot_item.sort_values(['Purchase Count'],ascending=False)
hot_item.head(1)
# revisit how to list in a table
```
