```python
import pandas as pd
import numpy as np
game_file = "../Resources/Purchase_data.csv"
```

```python
game_df=pd.read_csv(game_file)
game_df.head()
```

```python
u_player=game_df['SN'].nunique()
avg_price=game_df['Price'].mean()
total_purchase=len(game_df['Purchase ID'])
total_rev=game_df['Price'].sum()
```

```python
sum_game={'Unique Player Count':u_player, 'Average Purchase Price':avg_price,'Number of Purchase':total_purchase,'Total Revenue':total_rev}
sum_game
summary_df=pd.DataFrame([sum_game])
summary_df
```

```python
group_gender=game_df.groupby('Gender')
gender_count=group_gender['SN'].nunique()
gen_percent=gender_count/u_player*100
sum_gender=pd.DataFrame({'Gender Count':gender_count,'Gender Percentage':gen_percent})
sum_gender
count_purchase=group_gender['Purchase ID'].count()
avg_pur_price=group_gender['Price'].mean()
avg_total_pur=group_gender['Price'].sum()
avg_pur_person=avg_total_pur/gender_count*100
gender_pur_sum=pd.DataFrame({'Purchase Count':count_purchase,'Average Purchase Price':avg_pur_price,'Average Purchase':avg_total_pur,'Average Purchase Per Person': avg_pur_person})
gender_pur_sum
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

game_df['Age Group']=pd.cut(game_df['Age'],bins,labels=labels)
game_df
group_age=game_df.groupby('Age Group')
count_age=group_age['SN'].nunique()
percent_age=count_age/u_player*100
age_sum=pd.DataFrame({'Players in Age Group':count_age,'Percent of Age Group':percent_age})
age_sum
```
